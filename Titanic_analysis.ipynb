```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
train = pd.read_csv("train.csv")
```

```python
train.describe()
```

```python
train.info()
train.head()
```

```python
print(train.isnull().sum())

data_age_nan = train[train["Age"].isnull()]
info = data_age_nan.groupby("Pclass").count()
info
```

```python
from matplotlib import pyplot as plt

#Age

survived = train[train["Survived"] == 1]
not_survived = train[train["Survived"] == 0]

plt.hist(survived["Age"], label='Age survived', histtype = 'step')
plt.hist(not_survived["Age"], label='Age not survived', histtype = 'step')
plt.xlabel("Age")
plt.ylabel("No. of people")
plt.legend()
plt.title("Survived vs. not survived: Age")
plt.show()
```

```python
#Age of Men who survived/not survived

male_survived = survived[survived["Sex"] == "male"]
male_not_survived = not_survived[not_survived["Sex"] == "male"]



fig, ax = plt.subplots()
rects1 = ax.hist(male_survived["Age"], label='Male Survived', histtype = 'step')
rects2 = ax.hist(male_not_survived["Age"], label='Male Not Survived', histtype = 'step')
plt.xlabel("Age")
plt.ylabel("No. of people")
plt.legend()
plt.title("Survived vs. not survived: Male Age")
plt.show()
```

```python
#Age of Women who survived/not survived

female_survived = survived[survived["Sex"] == "female"]
female_not_survived = not_survived[not_survived["Sex"] == "female"]



fig, ax = plt.subplots()
rects1 = ax.hist(female_survived["Age"], label='Female Survived', histtype = 'step')
rects2 = ax.hist(female_not_survived["Age"], label='Female Not Survived', histtype = 'step')
plt.xlabel("Age")
plt.ylabel("No. of people")
plt.legend()
plt.title("Survived vs. not survived: Female Age")
plt.show()
```

```python
import seaborn as sns

#Survived vs. not survived: Pclass

pclass = train.groupby(["Survived", "Pclass"])
pclass.head()
```

```python
#Pclass
sns.histplot(data=train, x="Pclass", hue="Survived", element="step")
plt.title("Survived vs. not survived: Pclass")
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
#Survived vs. not survived in pclass: # of parents / children aboard the Titanic	
sns.set_palette("Spectral", n_colors=2)
g = sns.FacetGrid(train, col="Pclass", hue="Survived")
g.map(sns.histplot, "Parch", element="step")
plt.legend(["Not Survived", "Survived"])
```
